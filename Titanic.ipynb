## 数据预处理

```python
from sklearn.ensemble import RandomForestRegressor

##使用RandomForesClassifier 填补缺失的年龄属性
def set_missing_ages(df):
    #把已有的数值型特征取出来丢进 Random Forest Regressor 中
    age_df=df[['Age','Fare','Parch','SibSp','Pclass']]
    
    #乘客分成已知年龄和未知年龄俩部分
    known_age = age_df[age_df.Age.notnull()].as_matrix()
    unknown_ade = age_df[age_df.Age.isnull()].as_matrix()
    
    # y即目标年龄
    y = known_age[:, 0]
    
    #X 即特征属性值
    X = known_age[:, 1:]
    
    #fit到RandomForestRegressor之中
    rfr = RandomForestRegressor(random_state=0,n_estimators=2000,n_jobs=-1)
    rfr.fit(X,y)
    
    #用得到的模型进行未知年龄结果预测
    predictedAges = rfr.predict(unknown_ade[:, 1::])
    
    #用得到的预测结果值填补原缺失数据
    df.loc[(df.Age.isnull()),'Age'] = predictedAges
    
    return df,rfr
    
```

```python
#对有无舱位信息替换为yes和no
def set_Cabin_type(df):
    df.loc[(df.Cabin.isnull()),'Cabin']='No'
    df.loc[(df.Cabin.notnull()),'Cabin']='Yes'
    return df
```

```python
data_train,rfr=set_missing_ages(data_train)
data_train=set_Cabin_type(data_train)
```

```python
data_train
```

```python
dummies_Cabin = pd.get_dummies(data_train.Cabin,prefix='Cabin')

dummies_Embarked = pd.get_dummies(data_train.Embarked,prefix='Embarked')

dummies_Sex = pd.get_dummies(data_train.Sex,prefix='Sex')

dummies_Pclass = pd.get_dummies(data_train.Pclass,prefix='Pclass')

df = pd.concat([data_train,dummies_Cabin,dummies_Embarked,dummies_Sex,dummies_Pclass],axis=1)
df.drop(['Cabin','Embarked','Name','Sex','Pclass','Ticket'],axis=1,inplace=True)
```

```python
import sklearn.preprocessing as preprocessing
scaler = preprocessing.StandardScaler()
age_scale_param = scaler.fit(df['Age'])
df['Age_scaled'] = scaler.fit_transform(df['Age'],age_scale_param)
fare_scale_param = scaler.fit(df['Fare'])
df['Fare_scaled'] = scaler.fit_transform(df['Fare'],fare_scale_param)
```

## 逻辑回归建模


```python
from sklearn import linear_model

#用正则取出我们要的属性值
train_df = df.filter(regex='Survived|Age_.*|SibSp|Parch|Fare_.*|Cabin_.*|Embarked_.*|Sex_.*|Pclass_.*')
train_df.to_csv('dtrain.csv',index=False)
train_np = train_df.as_matrix()

#y即Survived结果
y = train_np[:, 0]

#X即特征属性值
X = train_np[:, 1:]

#fit到RandomFroestRegressor之中
clf = linear_model.LogisticRegression(C=1.0,penalty='l1',tol=1e-6)
clf.fit(X, y)

clf
```

```python
data_test = pd.read_csv('test.csv')
data_test.loc[(data_test.Fare.isnull(),'Fare')]=0
#接着对测试集做和训练集相同的处理
#首先用同样的RandomforestRegressor模型填上丢失的年龄
tmp_df = data_test[['Age','Fare','Parch','SibSp','Pclass']]
null_age = tmp_df[data_test.Age.isnull()].as_matrix()
#根据特征属性X预测年龄并补上
X = null_age[:, 1:]
predictedAges= rfr.predict(X)
data_test.loc[(data_test.Age.isnull()),'Age']=predictedAges

data_test=set_Cabin_type(data_test)
#进行one-hot编码
dummies_Cabin=pd.get_dummies(data_test.Cabin,prefix='Cabin')
dummies_Embarked=pd.get_dummies(data_test.Embarked,prefix='Embarked')
dummies_Sex=pd.get_dummies(data_test.Sex,prefix='Sex')
dummies_Pclass=pd.get_dummies(data_test.Pclass,prefix='Pclass')


df_test=pd.concat([data_test,dummies_Cabin,dummies_Embarked,dummies_Sex,dummies_Pclass],axis=1)
df_test.drop(['Name','Pclass','Embarked','Sex','Cabin','Ticket'],axis=1,inplace=True)
df_test['Age_scaled']=scaler.fit_transform(df_test.Age,age_scale_param)
df_test['Fare_scaled']=scaler.fit_transform(df_test.Fare,fare_scale_param)
df_test.to_csv('titanic_test.csv',index=False)
```

```python
test = df_test.filter(regex='Age_.*|SibSp|Parch|Fare_.*|Cabin_.*|Embarked_.*|Sex_.*|Pclass_.*')
test.to_csv('dtest.csv',index=False)
predictions= clf.predict(test)
result = pd.DataFrame({'PassengerId':data_test.PassengerId.as_matrix(),'Survived':predictions.astype(np.int32)})
result.to_csv('result.csv',index=False)
```

```python
def xgboost_predict(train_x,test,PassengerId):
    import xgboost as xgb
    from sklearn.cross_validation import train_test_split
    target='Survived'
    feature = [x for x in train_x.columns if x not in ['Survived']]
    train, val = train_test_split(train_x, test_size = 0.2,random_state=2017)
    y=train[target]
    val_y=val[target]
    train=train[feature]
    val=val[feature]
    dtest = xgb.DMatrix(test)
    dval = xgb.DMatrix(val,label=val_y)
    dtrain = xgb.DMatrix(train,label=y)
    params={
        'boosting':'gbtree',
        'objective':'binary:logistic',
        'early_stopping_rounds':100,
        'eval_metric':'auc',
        'gamma':0.3,
        'max_depth':12,
        'lambda':0.3,
        'subsample':0.7,
        'colsample_bytree':0.4,
        'min_child_weight':3,
        'eta':0.01,
        'nthread':7
        }
        
    watchlist = [(dval,'val'),(dtrain,'train')]
    xgboost_model = xgb.train(params,dtrain,num_boost_round=200,evals=watchlist)
    #xgboost_model.save_model('')
    xgboost_predict_y=xgboost_model.predict(dtest,ntree_limit=xgboost_model.best_ntree_limit)
    result = pd.DataFrame()
    result['PassengerId']=PassengerId
    result[target] = xgboost_predict_y
    
    return result
```

```python
train=pd.read_csv('dtrain.csv')
test=pd.read_csv('dtest.csv')
dtest=pd.read_csv('test.csv')
passagerId=dtest['PassengerId']
result=xgboost_predict(train,test,list(passagerId))
```
